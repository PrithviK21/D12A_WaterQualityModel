```python
dfmix = pd.read_excel("uh.xlsx")
dfmix
```

```python
for i in df.index:
    if df['PH'][i] > 14:
        stat_code = df['STATION CODE'][i]
        df['PH'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['PH MIN'].mean()
    if df['TEMPERATURE'][i] > 45:
        stat_code = df['STATION CODE'][i]
        df['TEMPERATURE'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['TEMPERATURE MIN'].mean()
    if df['CONDUCTIVITY'][i] > 1500:
        stat_code = df['STATION CODE'][i]
        df['CONDUCTIVITY'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['CONDUCTIVITY MIN'].mean()
    if df['DISSOLVED OXYGEN'][i] > 3:
        stat_code = df['STATION CODE'][i]
        df['DISSOLVED OXYGEN'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['DISSOLVED OXYGEN MIN'].mean()
    if df['BOD'][i] > 30:
        stat_code = df['STATION CODE'][i]
        df['BOD'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['BOD MIN'].mean()
    if df['NITRATE'][i] > 10:
        stat_code = df['STATION CODE'][i]
        df['NITRATE'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['NITRATE MIN'].mean()
    if df['TOTAL COLIFORM'][i] > 30000:
        stat_code = df['STATION CODE'][i]
        df['TOTAL COLIFORM'][i] = '%.2f' % dfmix[dfmix['STATION CODE'] == stat_code]['TOTAL COLIFORM MIN'].mean()
        
df.to_excel("Inconsistencies.xlsx")
```
